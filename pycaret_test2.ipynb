```python
import re
def remove_tags(text):
    """Remove html tags from a string"""
    text = re.sub(r'&#39', '', text)
    clean = re.compile('<.*?>')
    cleaner = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    text = re.sub(cleaner, ' ', text)
    # anything else
    text = re.sub(r'User requested priority of P\d (Easy|Medium|Hard|High|Low)', '', text)
    text = re.sub('\s\s+', ' ', text);
    # ignore any text following the word resolution case insensitively, use https://regexr.com/ to check exps
    text = re.sub('(?i)resolution.*', '', text)
    return text

stop_list=['hello', 'pete', 'kind', 'regards', 'cyrille', 'stan', 'adams', 'email', 'address', 'need', 'use', 
'problem', 'able', 'getting', 'values', 'used', 'dear', 'client', 'unable', 'even', 'though', 'how', 'kindly', 
'event', 'events', 'request', 'send', 'file', 'issue', 'project', 'attached', 'list', 'using', 'work', 'one', 'issues', 
'tried', 'trying', 'error', 'solution', 'errors', 'message', 'thank', 'know', 'meeting', 'first', 'name', 'last', 'fred', 
'we', 'start', 'https', 'thanks', 'help', 'i', 'e', 'like', 'com', 'see', 'would', '2017', '2020', '2021', 'hi', 'good', 'day', 
'occur', 'started', 'customer', 'please', 'hello', 'best', 'regards', 'only', 'this', 'the', 'why', 'user', 'requested', 'priority']

import nltk
import string
stopwords = nltk.corpus.stopwords.words('english')
pd.set_option('display.max_colwidth', 100)
# Define a function to handle all data cleaning
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    text = "".join((char if char.isalpha() or char.isnumeric() or char=='.' else " ") for char in text).split()
    #text = "".join((char if char.isalpha() or char.isnumeric() else " ") for char in text).split()
    #tokens = re.split('\W+', text)
    #tokens=' '.join(tokens)
    #text = "".join([word.lower() for word in text.split()])
    text = [word for word in text if (word not in stopwords and word not in stop_list and len(word)<15 and len(word)>1)]
    return text
```
